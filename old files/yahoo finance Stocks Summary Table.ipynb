# yahoo finance Stocks Summary Table

```python
# Import Libraries
import time
from datetime import datetime
from splinter import Browser
from bs4 import BeautifulSoup as bs
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import pymongo
```

# CBA

#### Extract

```python
# start web browser
browser = webdriver.Chrome(ChromeDriverManager().install())

# get source code
browser.get("https://au.finance.yahoo.com/quote/CBA.AX?p=CBA.AX")
html = browser.page_source
time.sleep(2)
df = pd.read_html(html)

# close web browser
browser.close()

# Get each table from the table list
summary_table_1 = df[0]
summary_table_2 = df[1]
# Join the tables together
total_summary_table = summary_table_1.append(summary_table_2)
# Rename the columns
total_summary_table = total_summary_table.rename(columns = {0:'Summary Metric', 1:'Value'})
# Show result
total_summary_table
```

#### Transform

```python
# Remove all the rows without values based on any of the financial year columns
summary_table_new = total_summary_table[total_summary_table['Value']!=''].copy()

# Replace all fields containing '-' with 0
summary_table_new = summary_table_new.replace('-',0)

# Set the index to Summary Metric
summary_table_new = summary_table_new.set_index('Summary Metric')

# Make it a series
summary_table_cba_value = summary_table_new[['Value']].copy()
summary_table_cba_value
```

```python
summary_df = summary_table_cba_value.to_dict()
```

#### Load

```python
# Load Into Mongodb
import pymongo
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)

#Drop existing Database
client.Stocks_db.summary.drop()

#CreateDatabase
db = client.stocks_db
collection = db.items

cba_summary_report={}
cba_summary_report['CBA'] = summary_df
cba_summary_report
```

```python
client.Stocks_db.summary.insert_one(cba_summary_report)
```
