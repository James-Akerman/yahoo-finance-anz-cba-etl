## Setup

```python
# Import Libraries
import time
from datetime import datetime
from splinter import Browser
from bs4 import BeautifulSoup as bs
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import pymongo
```

## CBA

#### Extract

```python
# Read the tables in the HTML page
# Scrape the Data
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)

url = 'https://au.finance.yahoo.com/quote/CBA.AX/financials?p=CBA.AX'
browser.visit(url)

time.sleep(1)

html = browser.html
soup = bs(html, "html.parser")
```

```python
# Get table headers
income_table_headers = soup.find_all('div', class_='D(tbr) C($primaryColor)')
# Get table rows
income_table_rows = soup.find_all('div', class_='D(tbr) fi-row Bgc($hoverBgColor):h')
print(income_table_headers)
```

```python
# Get the table headers
table_headers_list = []
for header in income_table_headers:
    for span in header.find_all('span'):
        table_headers_list.append(span.text)
```

```python
table_headers_list
```

```python
# Get all the table rows
table_rows = []
row_list = []
for rows in income_table_rows:
    for fields in rows:
        row_list.append(fields.text)
    # Add row to table rows
    table_rows.append(row_list)
    # Reset the row list for the next row
    row_list = []
```

```python
# Quit the browser
browser.quit()
```

#### Transform

```python
# Convert the result into a DataFrame
income_statement = pd.DataFrame(table_rows, columns=table_headers_list)
income_statement
```

```python
#Find empty or NaN entry in Dataframe

missing_cols, missing_rows = (
    (income_statement.isnull().sum(x) | income_statement.eq('').sum(x))
    .loc[lambda x: x.gt(0)].index
    for x in (0, 1)
)
income_statement.loc[missing_rows, missing_cols]
```

```python
#income_table[income_table['ttm'] == ''].index
```

```python
# Delete row with index label '1' and set new index 
Income_statement_df = income_statement.drop([1]).set_index('Breakdown')
Income_statement_df
```

```python
# Replace all fields containing '-' with 0
Income_statement_df = Income_statement_df.replace('-',0)
Income_statement_df
```

```python
# Create Dataframe/Series per year
#2017-2018
Income_statement_df_2018 = Income_statement_df[['29/06/2018']].copy()
#2018-2019
Income_statement_df_2019 = Income_statement_df[['29/06/2019']].copy()
#2019-2020
Income_statement_df_2020 = Income_statement_df[['29/06/2020']].copy()
#2020-2021
Income_statement_df_2021 = Income_statement_df[['29/06/2021']].copy()
# TTM
Income_statement_df_ttm = Income_statement_df[['ttm']].copy()
```

```python
Income_statement_df_ttm
```

#### Load

```python
# Module used to connect Python with MongoDb
import pymongo
```

```python
# The default port used by MongoDB is 27017
# https://docs.mongodb.com/manual/reference/default-mongodb-port/
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)

#Delete existing Database before Loading
#client.Stocks_db.income_statement.drop()
```

```python
print(Income_statement_df_ttm.to_dict()['ttm'])
```

```python
#Convert dataframe into dict and insert to MongoDB
client.Stocks_db.income_statement_cba.insert_one(Income_statement_df_2018.to_dict())
client.Stocks_db.income_statement_cba.insert_one(Income_statement_df_2019.to_dict())
client.Stocks_db.income_statement_cba.insert_one(Income_statement_df_2020.to_dict())
client.Stocks_db.income_statement_cba.insert_one(Income_statement_df_2021.to_dict())
client.Stocks_db.income_statement_cba.insert_one(Income_statement_df_ttm.to_dict())
```

```python
#Convert dataframe into dict and group it into a one report/object

cba_income_statements={'29/06/2018':Income_statement_df_2018.to_dict()['29/06/2018'],
                              '29/06/2019':Income_statement_df_2019.to_dict()['29/06/2019'],
                               '29/06/2020':Income_statement_df_2020.to_dict()['29/06/2020'],
                              '29/06/2021':Income_statement_df_2021.to_dict()['29/06/2021'],
                                'ttm':Income_statement_df_ttm.to_dict()['ttm']}
cba_report={}
cba_report["CBA"] = cba_income_statements
```

```python
#Insert object into MongoDB
client.Stocks_db.income_statement.insert_one(cba_report)
```
